```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from scipy import stats

%matplotlib inline  
```

```python
# Load prediction fie
syn1 = np.loadtxt("./../results/B3501.true/syn1.pred", dtype=str)
syn2 = np.loadtxt("./../results/B3501.true/syn2.pred", dtype=str)
syn3 = np.loadtxt("./../results/B3501.true/syn3.pred", dtype=str)
syn4 = np.loadtxt("./../results/B3501.true/syn4.pred", dtype=str)

# Extract the elements
measure = [float(syn1[i][1]) for i in range(len(syn1))]
peptides = [syn1[i][0] for i in range(len(syn1))]
pred = list(zip([float(syn1[i][2]) for i in range(len(syn1))],
          [float(syn2[i][2]) for i in range(len(syn2))],
          [float(syn3[i][2]) for i in range(len(syn3))],
          [float(syn4[i][2]) for i in range(len(syn4))]))

# Calculate average prediction values
mean_pred = [np.mean(pred[i][0:4]) for i in range(len(pred))]
```

```python
mean_pred
```

```python
# Performance evaluation

# store training performance
eval_perf = pearsonr(np.asarray(measure), np.asarray(mean_pred))[0]

# PERFORMANCE REPORT
fig = plt.figure(figsize=(5, 5), dpi = 70)

plt.scatter(mean_pred, measure)
plt.ylabel("Target Value", fontsize=10);
plt.xlabel("Prediction Value", fontsize=10);

# print performance
print("# Prediction PCC:", np.round(eval_perf, 4))
```
